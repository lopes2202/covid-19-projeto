```python
import pandas as pd

tabela = pd.read_csv("covid19_statistics.csv")

tabela = tabela.drop(columns="Unnamed")
tabela = tabela.drop(columns="date")
tabela = tabela.drop(columns="time_hms")

display(tabela)
```

```python
# Passo 3: Corrigir os problemas da base de dados (Tratamento de Dados)

display(tabela.info())
tabela = tabela.dropna()
display(tabela.info())
```

```python

contagem = tabela['continent'].value_counts()
print(contagem)

# contagem = tabela['continent'].value_counts(normalize=True)
# print(contagem)
```

```python
# Passo 5: Análise detalhada, taxa de porcentagem de pessoas que morreram em cada continente
import plotly.express as px

grafico = px.pie(tabela, names="continent", values="deaths.total", title=f"Proporção de Mortes por continente")
grafico.update_traces(textinfo='percent+label')
grafico.show()
```

```python
# Taxa de letalidade em cada continente

tabela["letalidade"] = (tabela["deaths.total"] / tabela["cases.total"]) * 100 # criei outra label de letalidade

import plotly.express as px
grafico = px.bar(tabela, x="continent",  y="letalidade",  title="Taxa de Letalidade (%)",  color="letalidade")

grafico.show()


```

```python
grafico = px.pie(tabela, names="continent", values="population(k)", title=f"População por continente")
grafico.update_traces(textinfo='percent+label')
grafico.show()
```

```python
grafico = px.pie(tabela, names="continent", values="cases.critical", title=f"Casos criticos por continente")
grafico.update_traces(textinfo='percent+label')
grafico.show()
```
